```python
import pandas as pd
import numpy as np
import csv
```

```python
# read the provided CSV data into DataFrames
stations="hawaii_stations.csv"
weather="hawaii_measurements.csv"
stations_data=pd.read_csv(stations,encoding="iso-8859-1")
weather_data=pd.read_csv(weather,encoding="iso-8859-1")
stations_df=pd.DataFrame(stations_data)
weather_df=pd.DataFrame(weather_data)
```

```python
weather_df.count()
```

```python
# Checking for duplicates in the measurements DataFrame
weather_df.duplicated().sum()
```

```python
# Clean the NaN's from the precipitation column (convert to zero)
weather_df['prcp'] = weather_df['prcp'].fillna(0)
```

```python
# Checking for null values
stations_df.duplicated().sum()
```

```python
# Checking for NaNs
stations_df.isnull().values.sum()
```
